```python
import tkinter as tk
from tkinter import messagebox

questions = [
    {
        "question": "What is the capital of Tunisia?",
        "options": ["Paris", "Sousse", "Tunis", "Berlin"],
        "answer": ["Tunis"]
    },
    {
        "question": "Which is a traditional food in Tunisia?",
        "options": ["Pizza", "Pasta", "Couscous", "Hamburger"],
        "answer": ["Couscous"]
    },
    {
        "question": "What are the multiples of 80?",
        "options": ["15", "25", "20", "40"],
        "answer": ["20", "40"]
    },
    {
        "question": "How many students of GoMycose in Tunisia?",
        "options": ["20,000", "40,000", "30,000", "over30,000"],
        "answer": ["over30,000"]
    },
    {
        "question": "As of 18 June, how many people died in Palestine?",
        "options": ["over 57,800", "70,000", "60,000", "56,900"],
        "answer": ["over 57,800"] 
        
  
      },
    {   "question": "who is the winner of balon d'or 2013?",
        "options": ["Ronaldo ", "Mesii", "franck Ribery", "Neymar"],
        "answer": ["Ronaldo"] 
        
    
     },
    { 
        "question": "How many people live in the world in 2023?",
        "options": ["8,062 milliards", "8,100 milliards", "8,150 milliards", "8,177"],
        "answer": ["8,062"] 

     },  
    { 
        "question": "who is the winer of the word cup 2022?",
        "options": ["Argentine", "potugal", "spain", "france"],
        "answer": ["Argentine"]   
        


    }]
class QuizApp:
    def __init__(self, master):
        self.master = master
        master.title("üáπüá≥ Tunisia Quiz Challenge")
        master.geometry("750x550")
        master.configure(bg="#f0f8ff")

        self.q_index = 0
        self.score = 0
        self.global_time_left = 40
        self.timer_id = None

        self.title_label = tk.Label(master, text="üáπüá≥ Tunisia Quiz Challenge", font=("Helvetica", 24, "bold"), bg="#f0f8ff", fg="#1a1a1a")
        self.title_label.pack(pady=15)

        self.info_frame = tk.Frame(master, bg="#f0f8ff")
        self.info_frame.pack()

        self.score_label = tk.Label(self.info_frame, text="Score: 0", font=('Arial', 13), bg="#f0f8ff", fg="green")
        self.score_label.pack(side="left", padx=20)

        self.timer_label = tk.Label(self.info_frame, text="Time Left: 40s", font=('Arial', 13), bg="#f0f8ff", fg="red")
        self.timer_label.pack(side="right", padx=20)

        self.question_label = tk.Label(master, text="", wraplength=700, font=('Arial', 17, "bold"), bg="#f0f8ff", justify="center")
        self.question_label.pack(pady=20)

        self.option_frame = tk.Frame(master, bg="#f0f8ff")
        self.option_frame.pack(pady=10)

        self.vars = []
        self.option_buttons = []
        for _ in range(4):
            var = tk.IntVar()
            chk = tk.Checkbutton(self.option_frame, text="", variable=var, onvalue=1, offvalue=0,
                                 font=("Arial", 15), bg="#ffffff", fg="#000", padx=20, pady=10,
                                 relief="ridge", bd=2, width=40, anchor="w", highlightthickness=0)
            chk.pack(pady=8)
            self.vars.append(var)
            self.option_buttons.append(chk)

        self.progress_label = tk.Label(master, text="", bg="#f0f8ff", font=("Arial", 12))
        self.progress_label.pack(pady=5)

        self.submit_button = tk.Button(master, text="Submit Answer", command=self.check_answer,
                                       bg="#007acc", fg="white", font=("Arial", 14), padx=15, pady=8)
        self.submit_button.pack(pady=20)

        self.load_question()
        self.update_global_timer()

    def load_question(self):
        q = questions[self.q_index]
        self.question_label.config(text=q["question"])
        self.progress_label.config(text=f"Question {self.q_index + 1} of {len(questions)}")
        for i, option in enumerate(q["options"]):
            self.option_buttons[i].config(text=option)
            self.vars[i].set(0)

    def update_global_timer(self):
        self.timer_label.config(text=f"Time Left: {self.global_time_left}s")
        if self.global_time_left > 0:
            self.global_time_left -= 1
            self.timer_id = self.master.after(1000, self.update_global_timer)
        else:
            messagebox.showinfo("‚è∞ Time's up!", f"Time is over!\nYou scored {self.score} out of {len(questions)}.")
            self.master.destroy()

    def check_answer(self):
        q = questions[self.q_index]
        selected = [q["options"][i] for i, var in enumerate(self.vars) if var.get() == 1]
        if not selected:
            messagebox.showwarning("Wait!", "Please select at least one answer.")
            return
        if set(selected) == set(q["answer"]):
            self.score += 1
            self.score_label.config(text=f"Score: {self.score}")
        self.q_index += 1
        if self.q_index < len(questions):
            self.load_question()
        else:
            self.show_result()

    def show_result(self):
        if self.timer_id:
            self.master.after_cancel(self.timer_id)
        messagebox.showinfo("Quiz Completed", f"You scored {self.score} out of {len(questions)}!")
        self.master.destroy()

root = tk.Tk()
app = QuizApp(root) 
root.mainloop() 
```
